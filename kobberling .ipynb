## The following takes the default options from Kobberling's website to calculate the Prospect Theory Value of a Prospect and then compares it with another prospect
https://psych.fullerton.edu/mbirnbaum/calculators/cpt_calculator.htm
and 
http://prospect-theory.behaviouralfinance.net/cpt-calculator.php


```python
from obj_gambles import *
from copy import copy
```

```python
#below are the prospects and settings from the website above, I have deliberately ordered them slighly differently
a=lot([ -200, 50, -50,100],[0.25, 0.25, 0.25,0.25])   #This creates and instance of the lottery object,a 
a.alpha=.88;a.beta=.88;a.gama=[.61,1];a.delta=[.69,1];a.lamda=2.25;a.uform='power';a.wform='k&t' #This will set the parameters for a
a.summary(verbose=True)  #If verbose is False it will not give the definitions of the columns
```

```python
#A graph of the probability transformations for this instance can be obtained by
pltsize(3,5)
a.graphp()
```

```python
#A graph of the value function is:
a.graphu(-50,50)
```

```python
#The following quantities can be recalled
print('The orginal probability and payoffs were', a.p,a.x)
print('The reordered probability and payoffs were', a.P,a.X)
print('The transformed probabilities are (after reodering)', a.W())
print('The utilities of the payoffs are (after reordering)', a.U())

print('The other key quantities already described in the summary are', a.u_eu(),a.u_pt(), a.ce_eu(),a.ce_pt())
```

```python
#If you want to compare the original (reordered) proabilities with their transfromed versions us
df=lookup(a.P,a.W())
df['diff']=df[2]-df[1]
df['diff'].plot(kind='bar',grid=True, title='W-P')
df
```

```python
#Let us copy the instance a, but then change the probability worping
b=copy(a)
b.wform='prelecII'
b.summary()  #If verbose is False it will not give the definitions of the columns
```

```python
#A graph of the probability transformations for this instance can be obtained by
pltsize(3,5)
a.graphp(); b.graphp()
```

```python
#If you want to compare the  two, to see how they change, including weather one is first of second order dominant
compare(b,a)
```
